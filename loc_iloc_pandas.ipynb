```python
#program on loc and iloc in pandas uasge
#Importing pandas and numpy
import pandas as pd
import numpy as np

# create a simple dataframe
data=pd.DataFrame({'age':[10,22,13,21,12,11,17],'section' : ['A','B','C','B','B','A','A'],'city':['Gurgon','Delhi','Mumbai','delhi','Mumbai','delhi','Mumbai'],'gender' : ['M','F','F','M','M','M','F'],'favourite_color' : ['red',np.NAN,'yellow',np.NAN,'black','green','red']})
#view the data
print(data)
```

```python
#find outhe vale of age is >=15
data.loc[data.age>=15]
```

```python
#finding out the all rows with morethan one condition
data.loc[(data.age>=12 ) & (data.gender== 'M')]
```

```python
#finding out the all rows with morethan one condition
data.loc[(data.age>=12 ) & (data.gender== 'F')]
```

**we can also select a range of rows using loc**
# using loc can also slice pandas dataframe over a range of indices.If the indices are not in sorted order
#then it will select only rows with index 1 and 3(see in below ex).
#if indices are not numbers then we can not slice our dataframe.
#in that case need to use "iloc" function to slice our pandas dataframe.


```python
#slice
print(data)
data.loc[1:3]
```

```python
#select only the required columns with a condition
data.loc[(data.age >= 12), ['city','gender']]
```

```python
#update the values of a particular column on selected rows
#example if value of age is >=12 then update vales of the column section to "M"

```

We can do this by running a for loop as well but if our dataset is big in size, then it would take forever to complete the task. Using loc in Pandas, we can do this within seconds, even on bigger datasets!

```python
#updatea  acolumn with a condition
data.loc[(data.age>=12),['section']]='M'
data
```

 **Update the values of multiple columns on selected rows**

```python
data.loc[(data.age>=20),['section','city']]=['S','pune']
data
```

**Select rows with indices using iloc**

while using iloc, need to specify the rows and colms by their integer indexes. if we wan tot slect only first and 3rd rows just simply need to put this index in to a list in the iloc statement with our dataframe.

```python
data.iloc[[1,3]]
#both are different      data.iloc[1:3]---it is selecting a range of rows but above command is selecting a rows
```

```python
data.iloc[[0,2]]
```

**Select rows with particular indices and particular columns**

earlier using the loc function we have selected a few columns , similarly the data columns can be selected using iloc also but we can provide the **index numbe rof the column instead of column name**

```python
data.iloc[[0,1,2],[0,2,3,4]]
```

**Select a range of rows using iloc**

We can slice a dataframe using iloc as well. We need to provide the start_index and end_index+1 to slice a given dataframe. If the indices are not the sorted numbers even then it will select the starting_index row number up to the end_index:

```python
data.iloc[1:3]
```

```python
data.iloc[0:3]
```

```python
# select a range of rows and columns using iloc using indexes range in list
data.iloc[0:3, 1:2]
```
